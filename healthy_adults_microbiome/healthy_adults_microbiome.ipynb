```r
mydata = sampleMetadata %>% filter (disease == "healthy" & age_category == "adult") %>% select(where(~ !base::all(base::is.na(.x))))
glimpse(mydata)
```

```r
metadata = mydata %>% select(study_name, sample_id, subject_id, body_site, age, age_category, gender, country, BMI)
head(metadata)
```

```r
write_tsv(metadata, "healthy_adults_metadata.tsv")
```

```r
abundance = mydata %>% returnSamples("relative_abundance")
abundance
```

```r
mat = assay(abundance, "relative_abundance")
head(mat)
```

```r
mat2 = data.frame(taxa = row.names(mat), mat)
head(mat2)
```

```r
write_tsv(mat2, "healthy_adults_microbiome.tsv")
```
